### Gráficos con Seaborn

```
import seaborn as sns
import matplotlib.pyplot as plt
```

### Propiedades del Sistema Coordenado

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

### Relaciones Estadísticas 

```
plt.figure(figsize=(3,3))
sns.relplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.show()
```

```
sns.relplot(
    data=tips,x='total_bill', y='tip',hue='smoker',style='smoker'
)
plt.show()
```

```
sns.relplot(
    data=tips, x='total_bill', y='tip', hue='smoker', style='time'
)
```

```
sns.relplot(
    data=tips, x='total_bill', y='tip', hue='size'
)
plt.show()
```

```
sns.relplot(
    data=tips, x='total_bill', y='tip', hue='size', palette='ch:r=-.5,l=.75'
)
plt.show()
```

```
sns.relplot(
    data=tips, x='total_bill', y='tip', size='size'
)
plt.show()
```

```
sns.relplot(
    data=tips,x='total_bill', y='tip', size='size', sizes=(10,100)
)
plt.show()
```

```
fmri = sns.load_dataset("fmri")
fmri.head()
```

```
sns.relplot(data=fmri, x="timepoint", y="signal")
plt.show()
```

```
sns.relplot(data=fmri,x='timepoint', y='signal',kind='line')
plt.show()
```

```
sns.relplot(
    data=fmri,x='timepoint', y= 'signal', kind='line', errorbar= None
)
plt.show()
```

```
sns.relplot(data=fmri, x='timepoint',y='signal',kind='line',errorbar='sd')
plt.show()
```

```
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal",
    estimator=None, height=4, aspect=1.5
)
plt.show()
```

```
sns.relplot(
    data=fmri, x='timepoint', y='signal',kind='line', hue= 'event'
)
plt.show()
```

```
sns.relplot(
    data=fmri, x='timepoint', y='signal',kind='line',
    hue='region', style='event'
)
plt.show()
```

```
sns.relplot(
    data=fmri, x='timepoint', y='signal',kind='line',
    hue='region', style='event', markers=True
)
plt.show()
```

```
sns.relplot(
    data=fmri, x='timepoint', y='signal',kind='line',
    hue='event', style='event'
)
plt.show()
```

```
sns.relplot(
    data=fmri.query("event == 'stim'"), kind="line",
    x="timepoint", y="signal", hue="region",
    units="subject", estimator=None,
)
```

Multiples Gráficas

```
sns.relplot(
    data=tips, x='total_bill', y='tip', hue='smoker', col='time'
)
```

```
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", hue="subject",
    col="region", row="event", height=3,
    estimator=None
)
```

```
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```

### Visualización de Distribución de Datos

Histogramas Univariados  

```
sns.displot(data=tips,x='total_bill')
```

```
sns.displot(tips,x='total_bill', binwidth=10)
```

```
sns.displot(tips, x='total_bill', bins=5)
```

```
sns.displot(tips, x='size', bins=[1,2,3,4,5,6,7])
```

```
sns.displot(tips,x='size', discrete=True)
plt.show()
```

```
sns.displot(tips, x="day", shrink=.8)
plt.show()
```
