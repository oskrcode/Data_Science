## Visualización de Distribuciones de Datos

Los datos cuentan una historia y para entender su composición se utiliza gráficas distribuciones, estas permiten realizar análisis de manera visual, con ellas se puede responder preguntas como:    
- ¿Cuál es la forma general de la distribución de los datos?
- ¿Los datos están sesgados hacia la izquierda o hacia la derecha?
- ¿Cuál es el rango de los valores observados?
- ¿Cuáles son los valores atípicos o extremos en los datos?
- ¿Existe alguna tendencia o patrón en la distribución?
- ¿Qué porcentaje de los datos se encuentra dentro de un cierto rango o intervalo?
- ¿Los datos siguen alguna distribución particular, como una distribución normal o una distribución de Poisson?
- ¿Existen múltiples picos o modas en la distribución de los datos?


### Importar Recursos

```
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins")
```

### Histogramas Univariados

Los histogramas representan la frecuencia de un conjunto de datos continuos o discretos, Usando el sistema cartesiano para graficar el histograma, en el eje x se representa los subconjuntos de los datos y la frecuencia está indicada  en el eje y.

```
sns.displot(penguins, x="flipper_length_mm")
plt.show()
```

#### Tamaño del BIN

El subconjunto de datos representado que no es más que un intervalo de información, se denomina BIN

- Tamaño del Intervalo 

```
sns.displot(penguins, x="flipper_length_mm", binwidth = 6)
plt.show()
```

- Número de BIN

```
sns.displot(penguins, x="flipper_length_mm", bins=10)
```

#### Intervalos Faltantes 
Es muy común encontrar que una variable tome valores pequeños en cada uno de los BIN, esto hace que queden espacios entre cada uno de ellos dificultando la interpretación de los datos. 

```
tips = sns.load_dataset("tips")
sns.displot(tips, x="size")
plt.show()
```

- BIN precisos 

```
sns.displot(tips, x="size", bins=[1, 2, 3, 4, 5, 6, 7])
plt.show()
```

- Parámetro automático 

```
sns.displot(tips, x="size", discrete=True)
plt.show()
```

- Shrink 

```
sns.displot(tips, x="day", shrink=.4)
plt.show()
```

#### Agregar una variable 

Otra variable condiciona la visualización de los datos, es decir, muestra las frecuencias de las observaciones teniendo en cuenta el comportamiento de la nueva variable adicionada.  


```
sns.displot(penguins, x="flipper_length_mm", hue="species")
plt.show()
```

- Cambiar capas a step

```
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
plt.show()
```

- Mover vertical

```
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

- Mover horizontal

```
sns.displot(penguins, x="flipper_length_mm", hue="sex", multiple="dodge")
plt.show()
```

#### Dividir en subplots

Un gráfico puede ser presentado en diferentes subplots, esto se usa utilizando una grilla filas y columnas.


- Columnas

```
sns.displot(penguins, x="flipper_length_mm", col="sex")
```

- Filas

```
sns.displot(penguins, x="flipper_length_mm", row="sex")
```

### Histograma normalizados


#### Normalizar usando density
Esta normalización se hace de forma que el área total del histograma sea igual a 1.


```
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density")
```

```
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density", common_norm=False)
```

#### Normalizar usando probalility
Normaliza teniendo en cuenta que las alturas de las barras sumen 1



```
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="probability")
```

#### Normalizar usando frequency
Normaliza teniendo en cuenta el ancho del BIN, es decir, muestra el número de observaciones dividido por el ancho del BIN 

```
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="frequency")
```
