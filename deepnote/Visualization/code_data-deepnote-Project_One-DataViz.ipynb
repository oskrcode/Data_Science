# Gráficas Usando Matplotlib

Matplotlib tiene dos interfaces de aplicación (APIs) o estilos de uso de la libreria

Interface implicita "pyplot"



## Recursos para Graficar


Python, a través de la biblioteca Matplotlib, permite realizar gráficos increíbles. El primer paso es importar la biblioteca, para ello se utiliza la siguiente instrucción:

```
import matplotlib.pyplot as plt
```

### Estructura Base para Graficas 

Recuerda que para realizar un gráfico es necesario contar con insumos básicos, en este caso, un conjunto de datos. Generalmente, estos datos se organizan en los ejes de abscisas y ordenadas para los gráficos en dos dimensiones. En este artículo, se utilizan dos vectores unidimensionales denominados "x" e "y". La estructura básica se muestra en la siguiente instrucción:


```
x= [1,2,3,4]
y= [1,2,3,4]
```

```
plt.plot(x,y)
plt.show()
```

### Multiples Graficas 

Las gráficas superpuestas son una herramienta poderosa en el análisis de datos, ya que permiten comparar múltiples conjuntos de información en un solo gráfico. Esta técnica es especialmente útil cuando se desea observar las relaciones y tendencias entre diferentes variables o grupos en un mismo contexto visual.

```
plt.plot(x,y)
plt.plot(x,[4,4,6,7])
plt.plot(x,[3,3,2,0])
plt.show()
```

### Tamaño de Lienzo

Un gráfico es una obra de arte que se utiliza para narrar una historia generada a partir de los datos. Sus propiedades ayudan a identificar y transmitir la idea deseada. La primera característica consiste en delimitar el lienzo en el cual se plasmará el gráfico. Para lograrlo, se emplea la siguiente instrucción:


```
plt.figure(figsize=(2,2))
plt.plot(x,y)
plt.show()
```

### Dividir el Lienzo

Cuando se tiene un conjunto de datos que genera múltiples gráficas y se pueden agrupar, la biblioteca permite dividir el lienzo en un sistema matricial de posiciones. Esto facilita la transmisión de información de manera más efectiva. Para lograrlo, se utilizan las siguientes instrucciones:


```
plt.subplot(2,1,1)
plt.subplot(2,1,2)
plt.show()
```

### Graficas Básicas

En un vasto océano de datos, que abarcan diversos tipos y provienen de múltiples fuentes, se encuentra codificada información valiosa. Decodificar esta información y extraer su mensaje requiere presentar de diversas formas. Para lograrlo, pyplot ofrece funciones que permiten crear diferentes tipos de gráficas. A continuación, se muestra la estructura de las gráficas más utilizadas.


```
plt.subplot(3,3,1)
plt.plot([1,2,3,4,5])
plt.subplot(3,3,2)
plt.plot([1,2,3,4,5],[4,5,6,7,8])
plt.subplot(3,3,3)
plt.scatter([1,2,3,5,6],[4,5,6,7,8])
plt.subplot(3,3,4)
plt.bar([1,2,3,5,6],[4,5,6,7,8])
plt.subplot(3,3,5)
plt.stem([1,2,3,5,6],[4,5,6,7,8])
plt.subplot(3,3,6)
plt.step([1,2,3,5,6],[4,5,6,7,8])
plt.subplot(3,3,7)
plt.fill_between([1,2,3,5,6],[4,5,6,7,8],[5,6,7,8,9])
plt.subplot(3,3,8)
plt.stackplot([1,2,3,5,6],[4,5,6,7,8],[5,6,7,8,9])
plt.show()
```

### Gráficas para Análisis Estadístico

Los gráficos estadísticos son una herramienta invaluable para comunicar de manera efectiva la información numérica. Permiten visualizar patrones, tendencias y relaciones entre los datos, lo que facilita su comprensión y análisis. Los gráficos brindan una representación visual clara y concisa, lo que ayuda a resumir grandes cantidades de datos en una forma más accesible y fácil de interpretar. Además, los gráficos pueden destacar puntos clave, enfatizar diferencias y proporcionar una visión panorámica de los resultados.


```
plt.subplot(4,3,1)
plt.hist([1,1,1,3,3,3,5,5,4,4])
plt.subplot(4,3,2)
plt.hist([[1,1,1,2,2,3],[1,1,4,4]])
plt.subplot(4,3,3)
plt.boxplot([1,2,3,4,5])
plt.subplot(4,3,4)
plt.boxplot([[1,2,3,4],[2,5,6,3]])
plt.subplot(4,3,5)
plt.errorbar([2,4,6],[0,5,4],[2,1,2],fmt='o')
plt.subplot(4,3,6)
plt.violinplot([1,2,2,2,3,4,5])
plt.subplot(4,3,7)
plt.violinplot([[1,2,2,2,3,4,5],[1,2,3,3,3,3,4,5]])
plt.subplot(4,3,8)
plt.eventplot([1,1,1,1,1,2,2])
plt.subplot(4,3,9)
plt.hist2d([1,1,2,2,3,3],[1,2,2,2,3,3])
plt.subplot(4,3,10)
plt.hexbin([1,2,3,4],[1,2,3,4],gridsize=(10,10))
plt.subplot(4,3,11)
plt.pie([1,2,3,4])
plt.show()
```

## Customización de Gráficas

Un gráfico es la representación visual de los datos. La imagen muestra una gran cantidad de información, pero a veces puede faltar contexto. Además, ayuda a diferenciar el tipo de mensaje que se transmite o se desea dar a conocer. La personalización de la imagen facilita la comprensión de la historia que se quiere contar.


### Título 

Un título ayuda a centrar y segmentar el mensaje, captando la atención del observador y mejorando la interpretación de la gráfica. Para agregar un título, utiliza la siguiente estructura:


```
plt.figure(figsize=(2,2))
plt.plot([1,2,3,4],[4,5,6,7])
plt.xlabel('tiempo')
plt.ylabel('Amplitud')
plt.title('Gráfica')
plt.show()
```

### Nombre de Ejes

Dar un nombre a los ejes facilita al lector comprender lo que se está representando dentro del lienzo. Para lograrlo, se utiliza la siguiente instrucción:


### Leyenda

Utilizar gráficos superpuestos es muy común al representar datos, ya que permiten realizar comparaciones rápidas. Una etiqueta que proporcione más información sobre el contexto es útil para introducir las leyendas. Para agregar leyendas, utiliza la siguiente instrucción:


```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,1,2,3],label= 'lineal')
plt.plot([0,1,2,3],[0,1,4,9],label='Cuadrado' )
plt.xlabel('tiempo')
plt.ylabel('Amplitud')
plt.title('Funcion')
plt.legend()
plt.show()
```

### Texto

Agregar texto a un gráfico siempre ayuda a que el mensaje sea captado con rapidez y aumenta la retención dentro del contenido. Para agregar textos, utiliza la siguiente instrucción:


```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,1,2,3],label= 'lineal')
plt.xlabel('tiempo')
plt.ylabel('Amplitud')
plt.title('Funcion')
plt.text(2,0,'y=mx+b')
plt.legend()
plt.show()
```

```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,1,2,3],label= 'lineal')
plt.xlabel('tiempo')
plt.ylabel('Amplitud')
plt.title('Funcion')
plt.text(2,0,r'$y=\frac{1}{2}x+b$')
plt.legend()
plt.show()
```

### Anotaciones 

Con frecuencia, los datos presentan información que es variada o diferente entre sí. Para ejemplificar esto, se utilizan las anotaciones, las cuales se pueden agregar mediante la siguiente instrucción:


```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,1,2,3],label= 'lineal')
plt.xlabel('tiempo')
plt.ylabel('Amplitud')
plt.title('Funcion')
plt.text(2,0,r'$y=\frac{1}{2}x+b$')
plt.annotate('Punto XY', xy=(1,1),xytext=(2,1),arrowprops=dict(color='green',width=2, headwidth=8, headlength=5 ))
plt.legend()
plt.show()
```

### Ecuaciones en las Etiquetas 

Las ecuaciones son muy comunes para proporcionar información contextual sobre los datos. Para activar este tipo de información, se utiliza la nomenclatura LaTeX.


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time'+ ' '+r'$x$')
plt.ylabel('Amplitud ['+ r'$\frac{1}{2}x$'+']')
plt.title('función'+r'$y=\frac{1}{2}x+0$')
plt.show()
```

### Transparencia en Texto

Controlar la opacidad en los textos y gráficas ayuda a resaltar la relevancia de ciertos elementos en un gráfico. Este tipo de transparencia se logra mediante el control de:


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time'+ ' '+r'$x$', alpha = 0.3)
plt.ylabel('Amplitud ['+ r'$\frac{1}{2}x$'+']', alpha= 0.7)
plt.title('función'+r'$y=\frac{1}{2}x+0$', alpha= 0.9)
plt.show()
```

### Color de Texto

En una gráfica, el color desempeña un papel crucial. Aunque los ejes generalmente se representan en escala de grises, también se puede agregar color para resaltar los ejes o los textos, lo que llama la atención. Para agregar color, utiliza la siguiente instrucción:


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time'+ ' '+r'$x$', alpha = 0.9, color='blue')
plt.ylabel('Amplitud ['+ r'$\frac{1}{2}x$'+']', alpha= 0.7, color= 'red')
plt.title('función'+r'$y=\frac{1}{2}x+0$', c='g')
plt.show()
```

### Rotar Texto

 En ocasiones, las etiquetas, textos, etc., pueden ser demasiado grandes y superponerse entre sí, lo que dificulta su legibilidad. Para solucionarlo, es necesario rotarlas, y esto se logra utilizando la siguiente instrucción:


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time', rotation = 20)
plt.ylabel('Amplitud' , rotation=120 )
plt.title('función', rotation= 270)
plt.show()
```

### Espacio de Texto

Los espacios entre los ejes y las etiquetas son importantes para garantizar la legibilidad de la información y mejorar la comunicación de tu historia.


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time', labelpad=40)
plt.ylabel('Amplitud', labelpad=20)
plt.title('función')
plt.show()
```

### Movimiento horizontal y vertical

Desplazar las etiquetas de forma horizontal o vertical resulta útil para captar la atención del lector y lograr que el gráfico luzca más ordenado y pulido.


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time', loc='right')
plt.ylabel('Amplitud',loc='top')
plt.title('función', loc='left')
plt.show()
```

### Tipo, Tamaño y familia de letra

Un gráfico agradable debe contar con un tipo, tamaño y familia de letra legible. Existen numerosas familias y tipos de letra entre los cuales elegir. Además, el tamaño de la letra debe ser adecuado en relación al lienzo. Para modificar estas propiedades, utiliza la siguiente instrucción:


```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time', fontfamily='sans-serif')
plt.ylabel('Amplitud', fontfamily='serif')
plt.title('función', fontfamily='monospace')
plt.show()
```

```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time', fontstyle='normal')
plt.ylabel('Amplitud', fontstyle='italic' )
plt.title('función', fontstyle='oblique')
plt.show()
```

```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xlabel('time', fontsize= 10)
plt.ylabel('Amplitud', fontsize= 15)
plt.title('función', fontsize= 20)
plt.show()
```

### Marcas de Ejes

Las divisiones de los ejes en un gráfico desempeñan un papel fundamental en la visualización de datos. Estas divisiones, también conocidas como "ticks", son los marcadores que se colocan a lo largo de los ejes para representar valores específicos. Aunque pueden parecer pequeños detalles, las divisiones de los ejes son de gran importancia por varias razones.

```
plt.figure(figsize=(5,2.7))
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xticks([1,2,3], labels=['a','b','d'])
plt.yticks([0,0.75,1.5],['Bajo','Medio','Alto'],rotation=45)
plt.xlabel('time')
plt.ylabel('Amplitud')
plt.title('función')
plt.show()
```

### Escala de Ejes

Los ejes desempeñan un papel crucial en la lectura de una gráfica, ya que proporcionan contexto y permiten captar la idea codificada en los datos. Las diferentes escalas presentes en los ejes facilitan la comprensión del comportamiento a lo largo del tiempo.


```
plt.figure(figsize=(5,5))
plt.subplot(2,2,1)
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xscale('linear')
plt.title('Lineal')
plt.subplot(2,2,2)
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xscale('log')
plt.title('Log')
plt.subplot(2,2,3)
plt.plot([0,1,2,3],[0,0.5,1,1.5])
plt.xscale('symlog')
plt.title('Symlog')
plt.show()
```

### Estilos 

Un gráfico que narra una historia interesante dirige la atención hacia donde se encuentra el núcleo de la comunicación. Esto se logra mediante la modificación de ciertas propiedades de la visualización.


- Grosor de Linea

```
plt.figure(figsize=(2.7,2))
plt.plot([0,1,2,3],[0,0.5,1,1.5], linewidth=4)
plt.show()
```

- Estilos de Linea

```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,0.5,1,1.5],linestyle='dashed')
plt.plot([0,1,2,3],[0,1,4,9],linestyle='dashdot')
plt.show()
```

- Marcas

```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,0.5,1,1.5],marker='*')
plt.show()
```

- marcas con color

```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,0.5,1,1.5],marker='*', markeredgecolor='red')
plt.show()
```

- Tamaño Marcas interno y externo

```
plt.figure(figsize=(3,2))
plt.plot([0,1,2,3],[0,0.5,1,1.5],marker='*', markeredgecolor='red', markersize=40, markeredgewidth=10)
plt.show()
```
